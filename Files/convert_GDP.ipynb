```python
import pandas as pd
```

```python
encodings=['utf-8','latin1','iso-8859-1']
for encoding in encodings:
    try:
        df = pd.read_csv(r"",encoding=encoding)
        print(f'successful')
        break
    except UnicodeDecodeError:
        print(f'failed')
```

```python
df.head()
```

```python
df_melted = pd.melt(df, id_vars = ['Country'], var_name = 'Year', value_name = 'GDP(B)')
```

```python
df_melted.to_csv('GDP(B).csv',index=False)
```
